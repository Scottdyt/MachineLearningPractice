# use more training data to avoid overfit 

```python
Network2.run_network("overfit\OverfitMoreData.json",30,training_set_size = 50000) #epoch = 30, trainnig data[:50000],mini-bash =10,learning rate = 0.5
```

```python
Network2.make_plots("overfit\OverfitMoreData.json",30,training_cost_xmin =0,test_accuracy_xmin =0,training_set_size=50000) 
```

### 仅仅循环了三十次，训练集和测试集上的损失都在减小，测试集上准确率达到了95%,与训练集精度差距很小。
